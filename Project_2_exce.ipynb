```python
from google.colab import files
import pandas as pd
import io
import numpy as np
```

```python
uploaded = files.upload()
df = pd.read_csv(io.BytesIO(uploaded['Hospital Emergency Room Data.csv']))
df.head()
```

```python
df.info()
```

```python
df = df.drop(columns=["Patient Admission Flag.1"])
# one column is removed
```

```python
df["Patient Admission Flag"]=df["Patient Admission Flag"].replace({
    True:"Admitted",
    False:"Not Admitted"
})
```

```python
df.head()
```

```python
df["Patient Gender"]=df["Patient Gender"].replace({
      "M":"Male",
      "F":"Female"
    })
```

```python
df["Patient Name"] = df["Patient First Inital"] + ". " + df["Patient Last Name"]

```

```python
df = df.drop(columns="Patient First Inital")
df = df.drop(columns="Patient Last Name")
```

```python
df["Patient Satisfaction Score"] = df["Patient Satisfaction Score"].fillna(0).astype(int)
```

```python
df = df.dropna()
df = df.drop_duplicates()
```
